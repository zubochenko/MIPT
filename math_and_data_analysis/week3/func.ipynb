```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.linalg
from scipy import optimize
import math
```

```python
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
def h(x):
    return int(f(x))
```

```python
x = [i for i in range(0,31)]
y = [f(i) for i in x]
plt.plot(x,y)
```

```python
print ("%.2f" % scipy.optimize.minimize(f,2,method='BFGS').fun)
```

```python
print ("%.2f" % scipy.optimize.minimize(f,30,method='BFGS').fun)
```

```python
print ("%.2f" % scipy.optimize.differential_evolution(f,bounds=[(1,30)]).fun)
```

```python
x = [i for i in range(0,31)]
y = [h(i) for i in x]
plt.plot(x,y)
```

```python
print ("%.2f" % scipy.optimize.minimize(h,30,method='BFGS').fun)
```

```python
print ("%.2f" % scipy.optimize.differential_evolution(h,bounds=[(1,30)]).fun)
```
