```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts
%matplotlib inline
```

```python
tex = sts.truncexpon(5)
ex = tex.rvs(1000)
```

```python
x = np.linspace(0,10,100)
pdf = tex.pdf(x)
plt.plot(x, pdf, color = "red", label='pdf')
plt.hist(ex, density = True, color = "grey", label='histogram')
plt.legend()
plt.show()
```

```python
tpl = (5, 10, 50)

for i in tpl:
    lst = []
    for _ in range(1000):
        lst.append(tex.rvs(i).mean())
    plt.hist(lst, 120, density=True, range=[0, 6], color = "grey", label='histogram')
    var = np.sqrt(1 / i)
    x = np.linspace(-1, 6, 100)
    norm = sts.norm.pdf(x , 1 , var)
    plt.plot(x, norm, color = "red", label='pdf')
    plt.legend()
    plt.title("Sample size %i" %i)
    plt.ylabel('f(x)')
    plt.xlabel('x')
    plt.show()
```
