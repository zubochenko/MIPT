```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.linalg
from scipy import optimize
import math
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
def h(x):
    return int(f(x))
```

```python
x = [i for i in range(0,31)]
y = [f(i) for i in x]
plt.plot(x,y)
```

```python
x = [i for i in range(0,31)]
y = [h(i) for i in x]
plt.plot(x,y)
```

```python
A = np.array([[1,1], [1, 15]]) #1 степень
b = np.array([1,15])
b = f(b)
c = np.linalg.solve(A, b)
print (c)
x = np.arange(0, 15.1, 0.1)
y = sum([c[i]*(x**i) for i in range(len(c))])
plt.plot(x,y)


A = np.array([[1,1,1], [1, 8, 64], [1, 15, 225]]) #2 степень
b = np.array([1,8,15])
b = f(b)
c = np.linalg.solve(A, b)
print (c)
x = np.arange(0, 15.1, 0.1)
y = sum([c[i]*(x**i) for i in range(len(c))])
plt.plot(x,y)


A = np.array([[1,1,1,1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]]) #3 степень
b = np.array([1,4,10,15])
b = f(b)
c = np.linalg.solve(A, b)
print (c)
x = np.arange(0, 15.1, 0.1)
y = sum([c[i]*(x**i) for i in range(len(c))])
plt.plot(x,y)
plt.annotate('Многочлен третьей степени', xy=(11, 2.6), xytext=(8, 4),
            arrowprops=dict(facecolor='black', shrink=0.025),)
```
